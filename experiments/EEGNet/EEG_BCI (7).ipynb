```python
import numpy as np

for rec in raw_all:
    subj = rec["subject"]
    record = rec["record"]
    raw = rec["raw"]  # Get the Raw object

    # Access data, times, and channel names from the Raw object
    data = raw.get_data()
    times = raw.times
    ch_names = raw.ch_names

    sfreq = raw.info['sfreq']  # Get sampling frequency from info
    info = mne.create_info(ch_names=ch_names, sfreq=sfreq, ch_types="eeg")
    # raw = mne.io.RawArray(data, info, verbose=False) # Create a new RawArray to apply filters

    # Band-pass filter
    raw.filter(l_freq=0.5, h_freq=50.0, verbose=False)

    # === Plot PSD only for chosen sample ===
    if rec is sample_rec:
        print(f"\nðŸ”¹ Band-Pass Filter (0.5â€“50 Hz) â†’ {record} from {subj}")
        # Compute PSD from the filtered raw object
        psd_band, freqs = psd_array_welch(raw.get_data(), sfreq=sfreq, fmin=0, fmax=80, n_fft=2048)
        psd_mean_band = np.mean(psd_band, axis=0)

        plt.figure(figsize=(8, 4))
        plt.semilogy(freqs, psd_mean_band)
        plt.title(f"{record} â€” After Band-Pass (0.5â€“50 Hz)")
        plt.xlabel("Frequency (Hz)")
        plt.ylabel("Power (ÂµVÂ²/Hz)")
        plt.tight_layout()
        plt.show()

    # --- Save filtered data back into the dictionary ---
    rec["raw"] = raw # Update the raw object in the dictionary with the filtered data
```

youâ€™ll see the curve drop after 50 Hz.

```python
# === 3ï¸âƒ£ Notch Filter (50 Hz) for all data ===
for rec in raw_all:
    subj = rec["subject"]
    record = rec["record"]
    raw = rec["raw"] # Get the Raw object

    # Notch filter
    raw.notch_filter(freqs=[50,60], verbose=False)

    # --- Save filtered data back into the dictionary ---
    rec["raw"] = raw  # save filtered raw object

    # === Plot PSD only for chosen sample ===
    # Access data, times, and channel names from the Raw object for plotting
    data = raw.get_data()
    times = raw.times
    ch_names = raw.ch_names
    sfreq = raw.info['sfreq']

    if rec is sample_rec:
        print(f"\nðŸ”¹ Notch Filter (50 Hz) â†’ {record} from {subj}")
        psd_notch, freqs = psd_array_welch(data, sfreq=sfreq, fmin=0, fmax=50, n_fft=2048)
        psd_mean_notch = np.mean(psd_notch, axis=0)

        plt.figure(figsize=(8, 4))
        plt.semilogy(freqs, psd_mean_notch)
        plt.title(f"{record} â€” After Notch Filter (50 Hz)")
        plt.xlabel("Frequency (Hz)")
        plt.ylabel("Power (ÂµVÂ²/Hz)")
        plt.xlim(0, 60)
        plt.tight_layout()
        plt.show()
```

#Removing artifacts by ICA

ICA separates signals into statistically independent sources, which often correspond to:

Brain activity

Eye blinks

Muscle artifacts

Line noise

```python
import mne
from mne.preprocessing import ICA

for rec in raw_all:
    subj = rec["subject"]
    record = rec["record"]
    raw = rec["raw"]  # Preprocessed Raw object

    # Determine number of ICA components: channels minus bad channels
    n_ica_components = raw.info['nchan'] - len(raw.info['bads'])
    if n_ica_components <= 0:
        print(f"Skipping ICA for {subj} - {record}: Not enough channels ({n_ica_components})")
        continue

    # Fit ICA
    ica = ICA(n_components=n_ica_components, random_state=42)
    try:
        ica.fit(raw)
    except ValueError as e:
        print(f"Could not fit ICA for {subj} - {record}: {e}")
        continue

    # Apply ICA (reconstruct EEG from all components)
    ica.apply(raw , inplace=True)

    # Update the raw object in the dictionary
    rec["raw"] = raw

print("ICA applied to all recordings.")
```
